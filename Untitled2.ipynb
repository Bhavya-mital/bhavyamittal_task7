logistic regression

```python
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
x=df[['Age','Area Income','Male','Daily Time Spent on Site','Daily Internet Usage']]
y=df['Clicked on Ad']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.33,random_state=42)
```

training

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

prediction and model evaluation

```python
predictions=model.predict(x_test)
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(y_test,predictions))
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(predictions,y_test)
print(score)
```

hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
parameter={'penalty':['l1','l2','elasticnet'],'C':[0.5, 0.6, 0.7, 0.8],'max_iter':[100,200,300]}
```

```python
model_regressor=GridSearchCV(model,param_grid=parameter,scoring='accuracy',cv=5)
```

```python
model_regressor.fit(x_train,y_train)
```

```python
print(model_regressor.best_params_)
```
